```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/Data2_Train.csv')
test = pd.read_csv('data/Data2_Test.csv')
```

```python
#create a flag for Train and Test dataset
train['Type'] = 'Train'
test['Type'] = 'Test'

fullData = pd.concat([train, test],axis=0)

fullData.head()
```

```python
fullData['Mileage'] = fullData['Mileage'].str.replace(' km/kg', '')
fullData['Mileage'] = fullData['Mileage'].str.replace(' kmpl', '')
fullData['Engine'] = fullData['Engine'].str.replace(' CC', '')
fullData['Power'] = fullData['Power'].str.replace('null bhp', '999.99')
fullData['Power'] = fullData['Power'].str.replace(' bhp', '')
```

```python
target_feature = ['Price']
categorical_features = ['Location', 'Fuel_Type', 'Transmission', 'Owner_Type']
numerical_features = ['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power','Seats']
other_features = ['Type']
```

```python
fullData.isnull().any()
```

```python
#combined numerical and categorical features
num_cat_features = numerical_features + categorical_features
```

```python
#create a new feature for each feature having missing values with 'feature_NA'
#and flag the missing value with 1 and other with 0
for var in num_cat_features:
    if fullData[var].isnull().any() == True :
        fullData[var + '_NA'] = fullData[var].isnull() * 1
```

```python
fullData.head()
```

```python
fullData[numerical_features] = fullData[numerical_features].fillna(value=999.99)


fullData['Mileage'] = fullData['Mileage'].astype(str).astype(float)
fullData['Engine'] = fullData['Engine'].astype(str).astype(float)
fullData['Power'] = fullData['Power'].astype(str).astype(float)


#Impute missing values

#FOR NOW impute missing numerical observations with mean

#fullData[numerical_features] = fullData[numerical_features].fillna(fullData[numerical_features].mean(), inplace=True)

fullData[categorical_features] = fullData[categorical_features].fillna(value = -9999)
fullData[numerical_features] = fullData[numerical_features].fillna(value = 999.99)
```

```python
#create label encoders for categorical features

from sklearn.preprocessing import LabelEncoder

for var in categorical_features:
    number = LabelEncoder()
    fullData[var] = number.fit_transform(fullData[var].astype('str'))

fullData.head()
```

```python
train=fullData[fullData['Type']=='Train']
test=fullData[fullData['Type']=='Test']

#train['is_train'] = np.random.uniform(0, 1, len(train)) <= .75
#Train, Validate = train[train['is_train']==True], train[train['is_train']==False]

from sklearn.model_selection import train_test_split

#spliting into training ans cross validation
#X = train.loc[:,['Location','Year','Kilometers_Driven','Fuel_Type','Transmission','Owner_Type','Mileage','Engine','Power','Seats' ]]
#X = fullData.loc[:,['Location','Year','Kilometers_Driven','Fuel_Type','Transmission','Owner_Type','Mileage','Engine','Power','Seats' ]]
X = train.loc[:,num_cat_features]

x_train, x_cv, y_train, y_cv = train_test_split(X, train.Price)

x_train.isna().any()

```

```python
from sklearn.linear_model import LinearRegression

lreg = LinearRegression()

#training the model
lreg.fit(x_train, y_train)
```

```python
#predicting on cv
pred = lreg.predict(x_cv)
```

```python
from sklearn.metrics import mean_squared_error,r2_score


#calculate mse
mse = np.mean((pred - y_cv)**2)
print(mse)
mse

mean_squared_error(pred,y_cv)
```

```python
x_test=test[list(num_cat_features)]
x_test

final_status = lreg.predict(x_test)
final_status
test["Predicted_Price"]=final_status

test
```

```python
from sklearn.tree import DecisionTreeRegressor

decisionTreeRegressor = DecisionTreeRegressor(random_state=0, max_features=10)

decisionTreeRegressor.fit(x_train, y_train)

pred = decisionTreeRegressor.predict(x_cv)
mean_squared_error(pred,y_cv)
```

```python
from sklearn.ensemble import RandomForestRegressor

randomForestRegressor = RandomForestRegressor(n_estimators=10, max_depth=25, random_state=1)

randomForestRegressor.fit(x_train, y_train)

pred = randomForestRegressor.predict(x_cv)
mean_squared_error(pred,y_cv)

final_status = randomForestRegressor.predict(x_test)
final_status
test["Predicted_Price"]=final_status

test
```
